```python
item_profile = tf.placeholder(tf.float32, shape=[num_videos, None], name="user_des")
user_profile_tr = tf.placeholder(tf.float32, shape=[None, 1], name="user_des")
user_desire = tf.placeholder(tf.int32, shape=[1], name="user_des")
normalize_a = tf.nn.l2_normalize(item_profile, 1)        
normalize_b = tf.nn.l2_normalize((user_profile_tr),1)
cos_similarity =tf.reshape(tf.matmul(normalize_a, normalize_b), [1, -1])
top_vals_all ,top_all_indice = tf.nn.top_k(cos_similarity, num_videos, sorted=True)
top_vals_k ,top_k_indice = tf.nn.top_k(cos_similarity, top_k, sorted=True)
top_vals_r ,top_r_indice = tf.nn.top_k(cos_similarity, top_r, sorted=True)
top_vals_l ,top_l_indice = tf.nn.top_k(cos_similarity, top_l, sorted=True)
```

```python
with tf.name_scope('user_accuracy'):

    
    to_cast_r = tf.cast(tf.equal(user_desire, top_r_indice), tf.float32)
    user_accuracy_r = tf.reduce_sum(tf.reduce_sum(to_cast_r, 1))
    
    to_cast_l = tf.cast(tf.equal(user_desire, top_l_indice), tf.float32)
    user_accuracy_l = tf.reduce_sum(tf.reduce_sum(to_cast_l, 1))

    to_cast_k = tf.cast(tf.equal(user_desire, top_k_indice), tf.float32)
    user_accuracy_k = tf.reduce_sum(tf.reduce_sum(to_cast_k, 1))
       
with tf.name_scope('mean_reciprocal_rank_with_top_k'):
    to_cast = tf.cast(tf.equal(user_desire, top_all_indice), tf.float32)
    to_where = (tf.where(tf.equal(to_cast, tf.ones(shape=[tf.shape(to_cast)[0], 1]))))[:,1]
```
